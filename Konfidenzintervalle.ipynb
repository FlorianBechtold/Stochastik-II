# Konfidenzintervalle 

### Wir veranschaulichen das Konzept anhand von Hausaufgabe 4.1 (Blumen in zwei Farben)
### Beim Züchten einer Blumensorte ergeben sich rot blühende und weiß blühende Exemplare. Nach den Regeln der Vererbung von Merkmalen muß eine der beiden Farben ein dominantes Merkmal sein und mit einer Wahrscheinlichkeit von 3/4 auftreten. Bei einem Versuch erhält der Züchter 13 Pflanzen, die schlußendlich blühen. Mit welcher Wahrscheinlichkeit irrt sich der Züchter, wenn er die häufiger auftretende der beiden Farben für dominant hält?


```python
%%latex
Sei $X$ die Anzahl an Blumen, welche dominant blühen. Nach Aufgabenstellung ist $X$ binomial verteilt mit $n=13$ und $p=3/4$. Wir nehmen also $H_0=3/4$.  
```

```python
def plotbinom(n, p):
    k=np.arange(13)
    pk=[binom.pmf(z, n, p) for z in k]
    plt.xlabel('k')
    plt.ylabel('P(X=k)')
    plt.title(f'Gewichte der Poisson Verteilung zum Parameter {(n, p)}')
    plt.bar(k, pk)
    plt.show()
    return None
    
plotbinom(13, 0.75);
```

```python
%%latex
In diesem Szenario irrt sich der Züchter, falls die rezessiven Blumen häufiger auftauchen als die dominanten, also wenn gilt $X\leq 6$ . Wir berechnen 
\[
\mathbb{P}_{3/4}(X\leq 6)=\sum_{k=0}^6 {13 \choose k} 0.75^k (1-0.75)^{13-k}\simeq 0,024
    \]
Graphisch gesehen summieren wir die Höhe aller Balken im obigen Diagramm für $k\leq 6$.
```

```python
%%latex
Alternativ hätte man auf $Y$ definieren können als Anzahl der rezessiven Blumen. In diesem Fall wäre $Y$ binominalverteilt mit $n=13$ und $p=1/4$. Hier ist nun $H_0=1/4$. 
```

```python
def plotbinom(n, p):
    k=np.arange(13)
    pk=[binom.pmf(z, n, p) for z in k]
    plt.xlabel('k')
    plt.ylabel('P(Y=k)')
    plt.title(f'Gewichte der Poisson Verteilung zum Parameter {(n, p)}')
    plt.bar(k, pk)
    plt.show()
    return None
    
plotbinom(13, 0.25);
```

```python
%%latex
In diesem Szenario irrt sich der Züchter ebenfalls, falls die rezessiven Blumen häufiger auftauchen als die dominanten, also wenn gilt $Y\geq 6$ . Wir berechnen 
\[
\mathbb{P}_{1/4}(Y\geq 7)=\sum_{k=7}^{13} {13 \choose k} 0.25^k (1-0.25)^{13-k}\simeq 0,024
    \]
Graphisch gesehen summieren wir die Höhe aller Balken im obigen Diagramm für $k\geq 6$.
```
