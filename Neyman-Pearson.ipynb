```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters




mu0 = 0       # Mean under H0
mu1 = 4    # Mean under H1
sigma = 1     # Known standard deviation
alpha = 0.05  # Significance level

# Critical value for alpha
critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma)

error2=norm.cdf(critical_value, loc=mu1, scale=sigma)

# x values for plotting
x = np.linspace(-3, 5, 500)
pdf_H0 = norm.pdf(x, loc=mu0, scale=sigma)
pdf_H1 = norm.pdf(x, loc=mu1, scale=sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_H0, label=rf'$H_0: \mu = {mu0}$', color='blue')
plt.plot(x, pdf_H1, label=rf'$H_1: \mu = {mu1}$', color='red')
plt.fill_between(x, 0, pdf_H0, where=(x > critical_value), color='blue', alpha=0.3, label='Type I error region (α)')
plt.fill_between(x, 0, pdf_H1, where=(x > critical_value), color='red', alpha=0.3, label='Power region (1 - β)')

# Vertical line for critical value
plt.axvline(x=critical_value, color='black', linestyle='--', label=f'Critical value ≈ {critical_value:.3f}')

# Labels and legend
plt.title('Neyman–Pearson Lemma: Most Powerful Test')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
print(f'Wahrscheinlichkeit eines Fehlers zweiter Art ist {np.round(error2, 3)}')
```

```python
def neymanmean(mu1=5, alpha=0.05):



    mu0 = 0       # Mean under H0
    mu1 = mu1      # Mean under H1
    sigma = 1     # Known standard deviation
    alpha = alpha # Significance level

# Critical value for alpha
    critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma)

    error2=norm.cdf(critical_value, loc=mu1, scale=sigma)

# x values for plotting
    x = np.linspace(-3, 5, 500)
    pdf_H0 = norm.pdf(x, loc=mu0, scale=sigma)
    pdf_H1 = norm.pdf(x, loc=mu1, scale=sigma)

# Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_H0, label=r'$H_0: \mu = 0$', color='blue')
    plt.plot(x, pdf_H1, label=rf'$H_1: \mu = {mu1}$', color='red')
    plt.fill_between(x, 0, pdf_H0, where=(x > critical_value), color='blue', alpha=0.3, label='Type I error region (α)')
    plt.fill_between(x, 0, pdf_H1, where=(x > critical_value), color='red', alpha=0.3, label='Power region (1 - β)')

# Vertical line for critical value
    plt.axvline(x=critical_value, color='black', linestyle='--', label=f'Critical value ≈ {critical_value:.3f}')

# Labels and legend
    plt.title('Neyman–Pearson Lemma: Most Powerful Test')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    print(f'Wahrscheinlichkeit eines Fehlers zweiter Art ist {np.round(error2, 3)}')
    return None

neymanmean(mu1=2, alpha=0.05)
```
